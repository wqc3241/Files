```python
import numpy as np
import pandas as pd
import matplotlib as mpl
import matplotlib.pyplot as plt
```

```python
file = "C:/Users/wqc32/Downloads/Vehicle Sales Report.xlsx"
df_17 = pd.read_excel(file,sheet_name = 1)
df_17 = df_17.rename({'Unnamed: 0': 'Vehicle', 'Total Buy Price(inlcudes 25 wire fee)':'Total Buy'}, axis=1)
df_17 = df_17[['Vehicle','Date','Total Buy', 'Sell price','Profit']]
df_17 = df_17.dropna()
df_17
plt.scatter(df_17['Sell price'], df_17['Profit'], color = 'blue')
plt.show()
plt.boxplot(df_17['Profit'], autorange=True)
```
